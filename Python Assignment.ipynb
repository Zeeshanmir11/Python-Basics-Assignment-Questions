6. Why are lists mutable, but tuples are immutable
 >The difference between the mutability of lists and the immutability of tuples in Python stems from their design and intended use cases.
 Lists are designed to be dynamic collections of items. This means you can change their contents after they are created. You can add, remove, or modify elements while Tuples, on the other hand, are designed to be static collections. Once a tuple is created, its contents cannot be changed.

8.What are logical operators in Python ?
 >In Python, logical operators are used to combine conditional statements (i.e., expressions that evaluate to True or False) and to perform logical operations on them. The three primary logical operators in Python are:

>1. And (Logical AND)
The and operator returns True if both conditions are True. If either condition is False, it returns False.
It is used to combine two or more conditions that must all be true for the overall expression to be true.
>2. or (Logical OR)
The or operator returns True if at least one of the conditions is True. It only returns False when both conditions are False.
It is used when you want either of two conditions to be true for the overall expression to be true.
>3. not (Logical NOT)
The not operator is used to invert the Boolean value of the condition. If the condition is True, not will make it False, and if the condition is False, not will make it True.
It is commonly used to reverse the result of a condition.

10.What is the difference between implicit and explicit type casting ?
> The difference between implicit and explicit type casting in Python revolves around who initiates the conversion and how it is handled.

>1. Implicit Type Casting
Initiated by Python: Implicit type casting happens automatically by Python when it converts a smaller data type to a larger data type. It is also known as type coercion.
No user intervention: Python automatically handles the conversion without needing explicit instructions from the programmer. This is done when it makes sense to prevent data loss or to avoid errors.
Safe and automatic: Python performs the conversion in a way that doesn't cause any loss of data or precision (typically from a lower precision type to a higher precision type).

>2.Explicit Type Casting (Type Conversion)
Initiated by the programmer: Explicit type casting requires the programmer to manually convert one data type to another using built-in functions like int(), float(), str(), etc.
User intervention required: The programmer decides when and how to convert data types, typically when the programmer needs to change data types to fit the requirements of the program.

```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments ?
# Function to demonstrate mutable and immutable arguments
def demonstrate_mutability(mutable_arg, immutable_arg):
    print("Original mutable argument:", mutable_arg)
    print("Original immutable argument:", immutable_arg)

    # Modifying the mutable argument (e.g., a list)
    mutable_arg.append(100)
    print("Modified mutable argument:", mutable_arg)

    # Modifying the immutable argument (e.g., an integer)
    immutable_arg += 10
    print("Modified immutable argument:", immutable_arg)

# Creating a mutable argument (list) and an immutable argument (integer)
my_list = [1, 2, 3]
my_int = 5

# Calling the function with mutable and immutable arguments
demonstrate_mutability(my_list, my_int)

# Checking the arguments after function call
print("\nAfter function call:")
print("Original list after function call:", my_list)  # List will be modified
print("Original integer after function call:", my_int)  # Integer will remain unchanged
```

```python
# 9Write a Python program to convert user input from string to integer, float, and boolean types.

user_input = input("Enter a value: ")

# Convert to integer, float, and boolean
try:
    # Convert to integer
    int_value = int(user_input)
    print(f"Integer: {int_value}")
except ValueError:
    print("The input is not a valid integer.")

try:
    # Convert to float
    float_value = float(user_input)
    print(f"Float: {float_value}")
except ValueError:
    print("The input is not a valid float.")

# Convert to boolean
# An empty string is considered False, everything else is considered True
bool_value = bool(user_input)
print(f"Boolean: {bool_value}")
```

```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50.?

# Initialize the sum variable
sum_of_evens = 0

# Loop through numbers from 1 to 50
for number in range(1, 51):
    if number % 2 == 0:  # Check if the number is even
        sum_of_evens += number  # Add the even number to the sum

print("The sum of all even numbers between 1 and 50 is:", sum_of_evens)
```

```python
# 15  Write a Python program to calculate the factorial of a number provided by the user using a while loop.

# Function to calculate factorial using a while loop
def factorial(n):
    result = 1
    while n > 1:
        result *= n  # Multiply the current result by n
        n -= 1  # Decrease n by 1
    return result

# Get input from the user
number = int(input("Enter a number: "))

# Calculate factorial
fact = factorial(number)

# Display the result
print(f"The factorial of {number} is: {fact}")
```
